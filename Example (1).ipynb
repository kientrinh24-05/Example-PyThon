```
# Bài 3
Person = {
      "name":"QuocNamm",
      "age":28,
      "Salary":8000,
      "City":"Tuy Hoa"
  }

KeyRemove = ["Salary","City"]

def Delitem(Person,KeyRemove):
# Xóa các key trong KeyRemove ra khỏi Person
  for x in KeyRemove:
      if x in Person:
          Person.pop(x)

  print(Person)

Delitem(Person,KeyRemove)
```
